```python
from netCDF4 import Dataset
import numpy as np
from cartopy import crs,feature
import matplotlib.pyplot as plt
```

THis is an example to load HFR total currents from SIO Thredds server

```python
netcdf_data = Dataset('http://hfrnet-tds.ucsd.edu/thredds/dodsC/HFR/USWC/6km/hourly/RTV/HFRADAR_US_West_Coast_6km_Resolution_Hourly_RTV_best.ncd')
```

```python
lat=netcdf_data.variables['lat'][:]
lon=netcdf_data.variables['lon'][:]
time=netcdf_data.variables['time'][:]
latest_time_index=int(len(time))-5
```

```python
u=netcdf_data.variables['u'][latest_time_index,:,:];
v=netcdf_data.variables['v'][latest_time_index,:,:];
data=np.sqrt(u**2+v**2)
```

```python
print(np.nanmax(data))
```

```python
plt.figure(figsize=(12,12))
#m=Basemap(projection='merc',lat_ts=10,llcrnrlon=lon.min(), \
#  urcrnrlon=lon.max(),llcrnrlat=lat.min(),urcrnrlat=lat.max(), \
#  resolution='i',area_thresh=100)
m=Basemap(projection='merc',llcrnrlon=-121, \
  urcrnrlon=-117,llcrnrlat=32,urcrnrlat=34.5, \
  resolution='i',area_thresh=100)

# convert the lat/lon values to x/y projections.

x, y = m(*np.meshgrid(lon,lat))

# mask data
data_m = np.ma.masked_invalid(data)
u_m = np.ma.masked_invalid(u)
v_m = np.ma.masked_invalid(v)

# plot the field using the fast pcolormesh routine 
# set the colormap to jet.

#m.pcolormesh(x,y,data_m,shading='flat',cmap=plt.cm.jet)
m.quiver(x,y,u,v,data_m,cmap=plt.cm.jet,scale=12,width=.001)
#m.colorbar()

# Add a coastline and axis values.

m.drawcoastlines()
m.fillcontinents(color='lightgrey');
m.drawmapboundary(fill_color='lightblue');
#m.drawparallels(np.arange(-90.,120.,30.),labels=[1,0,0,0])
#m.drawmeridians(np.arange(-180.,180.,60.),labels=[0,0,0,1])

# Add a colorbar and title, and then show the plot.

plt.title('Surface Current Magnitude from SIO TDS');
#plt.show()

```
